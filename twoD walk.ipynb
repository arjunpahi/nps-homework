```python
import matplotlib.pyplot as plt
import seaborn as sns
import random as random
import numpy as np
import pandas as pd
%matplotlib inline
sns.set()
```

```python
logfile = open("two_D/log.txt",'w')
```

```python
'''initialize first coordinate'''
x = 0
y = 0

'''create list to hold x and y coordinates'''
X = [x]
Y = [y]

'''initialize t for iteration'''
t=1

'''set itration number'''
ITR = 10000

'''set 'isprint' boolean variable to print in logfile/terminal'''
isprint = True

'''control datapoints with while loop'''
while t < ITR:
    direction =  random.choice(['x','y'])
    
    '''condition-I'''
    if direction == 'x':
        r = random.choice([-1,1])
        x = X[t-1]+r
        y = Y[t-1]
        X.append(x) ; Y.append(y)
    elif direction == 'y':
        r = random.choice([-1,1])
        x = X[t-1]
        y = Y[t-1]+r
        X.append(x) ; Y.append(y)
    if isprint:
        #print("step: ",t,"random number is: ", r,\
        #                "coordinate is : ", x,y)
        
        logfile.write("step: " + str(t) + " next move towards " + str(direction)  + str(r) + \
                         " current position is : " + str(x) + ","+str(y))
        
        logfile.write("\n")
        
        
    t = t+1
    
logfile.close()
```

```python
Data = {"x":X,"y":Y}
```

```python
import json as json
with open("two_D/walk.json",'w') as f:
    json.dump(Data,f)
```
