# Q. A body has initial velocity 20 $ \frac{m}{s}$ is moving with constant acceleration 10 $ \frac{m}{s^2}$. draw the displacement vs time plot within 0 to 100sec.

consider the general equation of motion

s = ut + $ \frac{1}{2}$ a $ t^2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u=20
a=10
t=np.linspace(0,100,10000)
s=u*t+.05*a*t**2
plt.plot(t,s)
plt.xlabel("t")
plt.ylabel('s')
```
